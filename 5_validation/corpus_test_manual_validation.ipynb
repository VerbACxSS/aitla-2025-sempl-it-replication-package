## Setup

```python
import json
import pandas as pd
```

## Utils

```python
TEXTS = ['1_proofreading', '2_lex', '3_connectives', '4_expressions', '5_sentence_splitter', '6_nominalizations', '7_verbs', '8_sentence_reorganizer']
```

## Load datasets

```python
dfs = {TEXT:[] for TEXT in TEXTS}

for TEXT in TEXTS:
    dfs[TEXT] = pd.read_csv(f'./manual_validation/corpus_test/{TEXT}.csv')
```

## RESULTS

```python
d = []
for TEXT, df in dfs.items():
    d.append({
        'Corpus':                                           TEXT,
        'È stata applicata la regola di semplificazione?':  df[df['È stata applicata la regola di semplificazione?'].isin(["SI", "PARZIALMENTE"])].shape[0],
        'Sono state eseguite delle operazioni aggiuntive?': df[df['Sono state eseguite delle operazioni aggiuntive?'] == "NO"].shape[0],
        'Sono stati introdotti degli errori?':              df[df['Sono stati introdotti degli errori?'] == "NO"].shape[0],
    })

pd.DataFrame(d).head(10)
```
